## Data Visualization using NumPY and Matplotlib

*NumPY, Pandas and Matplotlib are the key modules for Data analysis and Visualization with Python.<br/>*

```python
import numpy as np

# creating a numpy array 
 
# using the array() function, it converts the lists and tuples to numpy array

a = np.array([1,2,3,4,5])
print(type(a),"\n",a.dtype)
print("\n")

# We can also specify the data type.
# Numpy data type eg: .int64, .float64, .complex64 etc

a = np.array([1,2,3,4,5], dtype= np.complex64) #can also specify it as just complex
print(a)
print("\n")
 
# To create a Null or Zero array
a = np.zeros((2,2)) # (2,2) specify the order of the matrix
print(a)
print("\n")

# To create a Unit Matrix
a = np.ones((3,4))
print(a)
print("\n")

# To create an Identity Matrix
a = np.identity(3)
print(a)
print("\n")

# To create an array with random values
a = np.empty((3,3))
print(a)
print("\n")

#To create an array with sequences of numbers
a = np.arange(0,5,1) #syntax is np.arange(start,end,incriment)
print(a)
print("\n")

# The arange function always returns a 1D array which can be coverted
a = np.arange(1,11,1).reshape(5,2)
print(a)
print("\n")

#To create an array with sequences of floats 
a = np.linspace(0,2,10).reshape(2,5) # 10 numbers from 0 to 2
print(a)
print("\n")
```

#### Doing basic Math using the NumPY array

```python
# In NumPY the axis=0 refers to the columns 
# axix=1 refers to the rows 

a = np.random.random((2,2))
b = np.random.random((2,2))
print("Matrix A:\n",a,"\n")
print("Matrix B:\n",b,"\n")
print("The sum of all the elements is:",a.sum(),"\n") # find the sum of all the elements in the array

row_sum = a.sum(axis=1)
print("Row Sum:",row_sum)

column_sum = a.sum(axis=0)
print("Column Sum:",column_sum)

# Dot Product

print("Dot Product:\n",np.dot(a,b))

#Transpose of a Matrix

print("Trnaspose of matrix a is:\n",a.T)

#Inverse of a Matrix
a=np.matrix(a)
print("Inverse of Matrix A:\n",a.I)
```

#### Indexing and Slicing in NumPY array

```python
a = np.random.rand(4,4)
print(a,"\n")
# This is different from list where the syntax is a[0][0]
print("Element at the index (0,0) is:",a[0,0],"\n") 

#slicing
#The indices start from zero

#The 1st zero represents the starting row index and 2nd zero the column index
print("1st column elements:\n",a[0:4,0],"\n")
print("1st column elements:\n",a[:,0],"\n") #Equivalent to the step above

print("2nd row 2nd column element:\n",a[1:2,1],"\n")
print("1st row elements:\n",a[0:1,:],"\n")
print("1st row and 2nd row elements:\n",a[0:2,:],"\n")
print("3rd row 3rd column elements:\n",a[0:4,2:3],"\n")
print("1st row and 2nd row elements:\n",a[0:2,...],"\n")


"""NumPy also allows you to write this using dots,The dots (...) represent as many colons
   as needed to produce a complete indexing tuple.For example, if x is an array with 5 axes, then
   x[1,2,...] is equivalent to x[1,2,:,:,:],
   x[...,3] to x[:,:,:,:,3] and
   x[4,...,5,:] to x[4,:,:,5,:]
"""

"Note:"

a=np.arange(6)
print(a)
a1 = a[0:3]
print(a1)
a1[0] = 10 #This also changes the original array
print(a1)
print(a,"\n")

a=np.arange(6)
print(a)
a1 = a[0:3].copy() #This makes sure that the original array remains unchanged
print(a1)
a1[0] = 10
print(a1)
print(a,"\n")

#Advanced indexing

#1st method using list as indices
a=np.arange(1,11,1)
print(a)
print(a[[0,1,2,3,4]],"\n") #Here this list [0,1,2,3,4] represents the indices

#2nd using a Boolean array
a=np.arange(1,6,1)
print(a)
b=[True,True,False,False,False]
print(a[b])
```

```python
#A simple project to demonstrate the working of NumPY

import numpy as np
import matplotlib.pyplot as pp
import seaborn
import requests
%matplotlib inline 

open('./Files/stations.txt','r')
print("The Data Wwill Be Of This Format Given Below")
print("Station Code, Latitude, Longitude, Elevation, Station Name, GSN, some number\\n")
print("Example:")
print(open('./Files/stations.txt','r').readlines()[2]) 
#GSN is a special tag and it wont be present for every station.
"""The GSN is a baseline network comprising a subset of about 1000 stations chosen mainly to give a fairly 
uniform spatial coverage from places where there are a good length and quality of data record."""
```

```python
stations = {}

for lines in open('./Files/stations.txt','r'):# loop through each line in the Stations.txt file
    if 'GSN' in lines:
        fields = lines.split() #Returns a list
        stations[fields[0]] = ' '.join(fields[4:])# set station code as the key for the dictionary.
        #Join the everything from station name to the end of the line with space inbetween.
print("No of Stations:",len(stations),"\n")

def findstation(s):
    found = {code: name for code,name in stations.items() if s in name}
    print(found)
    

#We will be using these 4 stations for this project.
findstation('LIHUE')
findstation('SAN DIEGO')
findstation('MINNEAPOLIS')
findstation('IRKUTSK')

#Storing the station codes

datastations = {'USW00022536','USW00023188','USW00014922','RSM00030710'}

#The information regarding how to make sense of the .dly file is in the readme file in Files folder.
print("\n")

 #The basics one needs to know before using the .dly files
open('./Files/readme.txt').readlines()[99:131] 
```

```python
#Function to parse the file
def parsefile(filename):
    return np.genfromtxt(filename,
                            delimiter = dly_delimiter,
                            usecols = dly_usecols,
                            dtype = dly_dtype,
                            names = dly_names)

#Define the arguments
dly_delimiter = [11,4,2,4] + [5,1,1,1] * 31
dly_usecols = [1,2,3] + [4*i for i in range(1,32)]
dly_dtype = [np.int32,np.int32,(np.str_,4)] + [np.int32] * 31
dly_names = ['year','month','obs'] + [str(day) for day in range(1,31+1)]

#Parse the file.
lihue = parsefile('./Files/USW00022536.dly')
lihue
```

```python
#Define function to unroll the parsed file.
def unroll(record):
    startdate = np.datetime64('{}-{:02}'.format(record['year'],record['month']))
    dates = np.arange(startdate,startdate + np.timedelta64(1,'M'),np.timedelta64(1,'D'))

    rows = [(date,record[str(i+1)]/10) for i,date in enumerate(dates)]

    return np.array(rows,dtype=[('date','M8[D]'),('value','d')])

#Now check the output after unroll.
print(unroll(lihue[0]))
```

```python
#Define getobs function to concatenate all records after parse and unroll.
def getobs(filename,obs):
    return np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])

#Check the output and compare it with previous exercises.
print(getobs('./Files/USW00022536.dly','TMIN'))
```

```python

#Take the output for minimum and maximum values.
lihue_tmax = getobs('./Files/USW00022536.dly','TMAX')
lihue_tmin = getobs('./Files/USW00022536.dly','TMIN')

#Plot the output
pp.plot(lihue_tmax['date'],lihue_tmax['value'])
pp.show()
```

```python
#Define getobs function to concatenate all records after parse and unroll.
def getobs(filename,obs):
    data = np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])

    data['value'][data['value'] == -999.9] = np.nan

    return data

#Take the output for minimum and maximum values.
lihue_tmax = getobs('./Files/USW00022536.dly','TMAX')
lihue_tmin = getobs('./Files/USW00022536.dly','TMIN')

#Plot the output
pp.plot(lihue_tmax['date'],lihue_tmax['value'])
pp.plot(lihue_tmin['date'],lihue_tmin['value'])
pp.show()
```

```python
#Write a function to fill NaN values by interpolation.
def fillnans(data):
    dates_float = data['date'].astype(np.float64)

    nan = np.isnan(data['value'])

    data['value'][nan] = np.interp(dates_float[nan],dates_float[~nan],data['value'][~nan])

#Fill NaN values for max and min records.
fillnans(lihue_tmax)
fillnans(lihue_tmin)

#Calculate mean
print(np.mean(lihue_tmin['value']), np.mean(lihue_tmax['value']))
```

```python
#Write a function to smooth the plot.
def plot_smoothed(t,win=10):
    smoothed = np.correlate(t['value'],np.ones(win)/win,'same')

    pp.plot(t['date'],smoothed)

#We will try a couple of different smoothing parameters and also compare with the plot without smoothing.
pp.plot(lihue_tmin[10000:12000]['date'],lihue_tmin[10000:12000]['value'])

plot_smoothed(lihue_tmin[10000:12000])
plot_smoothed(lihue_tmin[10000:12000],30)

#Show the plot
pp.show()
```

```python
pp.figure(figsize=(10,6))

for i,code in enumerate(datastations):
    pp.subplot(2,2,i+1)

    plot_smoothed(getobs('./Files/{}.dly'.format(code),'TMIN'),365)
    plot_smoothed(getobs('./Files/{}.dly'.format(code),'TMAX'),365)
    
    pp.title(stations[code])
    pp.axis(xmin=np.datetime64('1952'),xmax=np.datetime64('2012'),ymin=-10,ymax=30)

pp.tight_layout()

#Show the plot
pp.show()
```

```python
def selectyear(data,year):
    start = np.datetime64('{}'.format(year))
    end = start + np.timedelta64(1,'Y')

    return data[(data['date'] >= start) & (data['date'] < end)]['value']

selectyear(lihue_tmin,1951)
```

```python
lihue_tmin_all = np.vstack([selectyear(lihue_tmin,year)[:365] for year in range(1951,2014+1)])
print(lihue_tmin_all,'\n')
print(lihue_tmin_all.shape)
```

```python
lihue_tmin_recordmin = np.min(lihue_tmin_all,axis=0)
lihue_tmin_recordmax = np.max(lihue_tmin_all,axis=0)
pp.plot(lihue_tmin_recordmax,'.')
pp.show()
```

```python
#Arrange all the years in a matrix
lihue_tmin_all = np.vstack([selectyear(lihue_tmin,year)[:365] for year in range(1951,2014+1)])
lihue_tmax_all = np.vstack([selectyear(lihue_tmax,year)[:365] for year in range(1951,2014+1)])

days = np.arange(1,365+1)

pp.figure(figsize=(12,4))

pp.fill_between(days,np.min(lihue_tmin_all,axis=0),np.max(lihue_tmin_all,axis=0),alpha=0.4)
pp.plot(selectyear(lihue_tmin,2009))

pp.fill_between(days,np.min(lihue_tmax_all,axis=0),np.max(lihue_tmax_all,axis=0),alpha=0.4)
pp.plot(selectyear(lihue_tmax,2009))

pp.axis(xmax=365)
# pp.show()
```

```python
#Find the year with the highest mean TMAX in Minneapolis
#Find the year with the lowest mean TMIN in San Diego
#In the same plot, show TMIN and TMAX throughout those years for the city

minneapolis_tmin = getobs('./Files/USW00014922.dly','TMIN')
minneapolis_tmax = getobs('./Files/USW00014922.dly','TMAX')

San_diego_tmin = getobs('./Files/USW00023188.dly','TMIN')
San_diego_tmax = getobs('./Files/USW00023188.dly','TMAX')

fillnans(minneapolis_tmax)
fillnans(minneapolis_tmin)
fillnans(San_diego_tmax)
fillnans(San_diego_tmin)

years = np.arange(1940,2014+1)

minneapolis_tmax_all = np.vstack([selectyear(minneapolis_tmax,year)[:365] for year in years])

minneapolis_mean = np.mean(minneapolis_tmax_all,axis=1)
pp.plot(years,minneapolis_mean)
pp.show()
```

```python
minneapolis_warmest = years[np.argmax(minneapolis_mean)]
minneapolis_warmest
```

```python
San_diego_tmin_all = np.vstack([selectyear(San_diego_tmin,year)[:365] for year in years])
San_diego_mean = np.mean(San_diego_tmin_all,axis=1)
San_diego_coldest = years[np.argmin(San_diego_mean)]
San_diego_coldest
```

```python
days = np.arange(1,366+1)

pp.figure(figsize=(12,4))

pp.fill_between(days,
                selectyear(minneapolis_tmin,minneapolis_warmest),
                selectyear(minneapolis_tmax,minneapolis_warmest),
                color = 'b',alpha=0.4)
pp.fill_between(days,
                selectyear(San_diego_tmin,San_diego_coldest),
                selectyear(San_diego_tmax,San_diego_coldest),
                color = 'r',alpha=0.4)
pp.axis(xmax=366)

pp.title('{} in Minneapolis vs {} in San Deago'.format(minneapolis_warmest,San_diego_coldest))

pp.show()
```
