## <p style="text-align:center">Manipulating Images</p>

*Pillow is a third-party Python module for interacting with image files.In Pillow, RGBA values are represented by a tuple of four integer values where 'A' stands for the level of transparency.<br/>
Pillow offers the ImageColor.getcolor() function so you don’t have to memorize RGBA values for the colors you want to use. This function takes a color name string as its first argument, and the string 'RGBA' as its second argument, and it returns an RGBA tuple.*

```python
from PIL import ImageColor
from IPython.display import display 
print(ImageColor.getcolor('red', 'RGBA'))
print(ImageColor.getcolor('CornflowerBlue', 'RGBA'))
```

*Pillow’s functions and methods take a box tuple argument. This means Pillow is expecting a tuple of four integer coordinates that represent a rectangular region in an image ie of the form (x<sub>1</sub>,y<sub>1</sub>,x<sub>2</sub>,y<sub>2</sub>).*

- x<sub>1</sub> is the x-coordinate of the leftmost edge of the box.
- y<sub>1</sub> is the y-coordinate of the top edge of the box.
- x<sub>2</sub> is the x-coordinate of one pixel to the right of the rightmost edge of the box. 
- y <sub>2</sub> is the y-coordinate of one pixel lower than the bottom edge of the box.

*The Image.new() function, which returns an Image object—much like Image.open(), except the image represented by Image.new()’s object will be blank. The arguments to Image.new() are as follows:*

- The string 'RGBA', which sets the color mode to RGBA. (There are other modes that this book doesn’t go into.) 
- The size, as a two-integer tuple of the new image’s width and height.
- The background color that the image should start with, as a four-integer tuple of an RGBA value. You can use the return value of the ImageColor.getcolor() function for this argument. Alternatively, Image.new() also supports just passing the string of the standard color name.

```python
im = Image.new('RGBA', (100, 200), 'purple')
im.save('./Files/purpleImage.png')
im1=Image.open('./Files/purpleImage.png')
display(im1)
```

### **Cropping Images**

```python
croppedIm = catIm.crop((335, 345, 565, 560))
croppedIm.save('./Files/cropped.png')
CCIm=Image.open('./Files/cropped.png')
display(CCIm)
```

### **Copying and Pasting Images onto Other Images**

```python
catCopyIm = catIm.copy()
faceIm = catIm.crop((335, 345, 565, 560))
catCopyIm.paste(faceIm, (0, 0))
catCopyIm.paste(faceIm, (400, 500))
catCopyIm.save('./Files/pasted.png')
CIm=Image.open('./Files/pasted.png')
display(CIm)
```

```python
catImWidth, catImHeight = catIm.size
faceImWidth, faceImHeight = faceIm.size
catCopyTwo = catIm.copy()
for left in range(0, catImWidth, faceImWidth):
    for top in range(0, catImHeight, faceImHeight):
        print(left, top)
        catCopyTwo.paste(faceIm, (left, top))
catCopyTwo.save('./Files/tiled.png')
CCT=Image.open('./Files/tiled.png')
display(CCT)
```

### **Resizing an Image**

```python
width, height = catIm.size
quartersizedIm = catIm.resize((int(width / 2), int(height / 2)))
quartersizedIm.save('./Files/quartersized.png')
Qt=Image.open('./Files/quartersized.png')
display(Qt)
```

### **Rotating and Flipping Images**

```python
croppedIm.rotate(90).save('./Files/rotated90.png')
c9=Image.open('./Files/rotated90.png') 
display(c9)
```

```python
croppedIm.rotate(6, expand=True).save('./Files/rotated6_expanded.png')
c8=Image.open('./Files/rotated6_expanded.png')
display(c8)
```

```python
croppedIm.transpose(Image.FLIP_LEFT_RIGHT).save('./Files/horizontal_flip.png')
croppedIm.transpose(Image.FLIP_TOP_BOTTOM).save('./Files/vertical_flip.png')
CH=Image.open('./Files/horizontal_flip.png')
display(CH)
CH=Image.open('./Files/vertical_flip.png')
display(CH)
```
